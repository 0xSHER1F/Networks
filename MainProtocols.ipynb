# **OSI(Open Systems Interconnection) Model :**

 * Conceptual Framework That Standardizes The Functions Of A Telecommunication Or Computing System Into Seven Distinct Layers

 * **Layers :**

     1) **Application (7) :**

         * Provides Network Services Directly To End-Users Or Application Processes
         * Includes Protocols Such As SMTP, FTP And DNS

     2) **Presentation (6) :**

         * Responsible For Data Translation, En/Decryption, and Compression  
         * Ensures That Data Exchanged Between Applications Is In A Format That The Receiving Application Can Understand
         * Includes Protocols Such As JPG, GIF, HTTPS, SSL And TLS

     3) **Session (5) :**

         * Establishes, Manages, And Terminates Connections Between Applications   
         * Provides Synchronization And Checkpointing Functions To Allow For Error Recovery And Data Integrity  
         * Includes Protocols Such As NetBIOS And PPTP

     4) **Transport (4) :**

         * Ensures Reliable End-To-End Communication Between Hosts 
         * Provides Error Detection And Correction, Flow Control, And Congestion Avoidance
         * Includes Protocols Such As TCP And UDP 

     5) **Network (3) :**

         * Responsible For Routing And Forwarding Data Packets Between Different Networks      
         * Reads The IP Address(Logical Address) From The Packet 
         * Routers Operate In This Layer 

     6) **Data Link (2) :**

         - Provides Node-To-Node Connectivity And Data Transfer 
         - Reads The MAC Address From The Data Packet  
         - Switches Operate In This Layer    

     7) **Physical (1) :**

         * Concerned With The Transmission Of Raw Data Its Over The Physical Medium

# **How to Communicate with HTTPS Pages ?**

 1) Client’s Browser Requests Secure HTTPS Page
 
 2) When you connect to an HTTPS site, the Web server sends its SSL certificate to your browser And Public Key 
 
 3) Client’s Browser Checks Validity Of Certificate Then Creates Symmetric Session Key Using Public And Sends It To Web Server 
 
 4) Web Server Decrypts Symmetric Session Key Using Its Private Key And Sends Page Encrypted Using Symmetric Session Key 

 5) Client’s Browser And Web Server Can Now Communicate Securely By En/Decrypting Data With Symmetric Session Key

# **DHCP(Dynamic Host Configuration Protocol) :**

 * Automatically Assign IP Addresses And Other Configuration Parameters To Devices That Connect To The Network

 * **D-O-R-A process in DHCP :**

     1) DHCP**D**iscover :
          
         - a client sends out a broadcast message(to 255.255.255.255) on the local network.

         - This broadcast message, called a DHCP Discover packet, requests a DHCP server to provide configuration information.

     2) DHCP**O**ffer :

         - In Response To The DHCP Discover Packet, DHCP Servers Sends DHCP Offer Packets Aontain IP Addresses And Configuration Information That The Server Is Willing To Provide To The Client 

     3) DHCP**R**equest :

         * The Client Chooses An offer Then Sends A DHCP Request Packet, Indicating Its Acceptance Of The Offered IP Address And Associated Configuration Options  

     4) DHCP**A**cknowledge :

         * The DHCP Server Sends A DHCP Acknowledge Packet That Confirms The Allocation Of The Requested IP Address And Provides The Client With The Selected Configuration Parameters 


     * After Completing The D-O-R-A Process, The Client Is Configured With A Valid IP Address And Other Necessary Network Configuration Parameters

# **DNS (Domain Name System) :**

 * Translates Human-Readable Domain Names Into IP Addresses And vice Versa

 * To Display Various Info For Certain Device :

     - Use Command ==> ipconfing /all (Win) , ifconfig (Linux)

 * **nslookup (DomainName) :**

     - Tells The Device To Ask Its DNS Server The IP Address Of The Specific


 * **dig @NS example.com :**

     * To Perform A Query For A Specific DNS Server    


 * **mxtoolbox.com**

     * Web Site For DNS Records   

 * **DNSchef :**

     * DNS proxy tool acts as a fake DNS server on your machine

     * When applications or devices configured to use it make DNS requests, DNSchef intercepts them instead of querying the real DNS servers       


 * **Common Attacks :**

     * **DNS Spoofing :**

         * Attacker Intercepts DNS Queries And Responds With False Information, Redirecting Users To A Malicious Website  


     * **DNS Amplification :**

         * Attacker Sends A Small DNS Query To A DNS Server With The Source IP Address Spoofed To The Intended Victim's IP Address.
         
         * The DNS Server Responds To The Query With A Much Larger Response, Overwhelming The Victim's Network   

     * **DNS Cache Snooping :**

         * Attacker Tries To Obtain Information About The DNS Servers Used By A Network By Analyzing DNS Queries And Responses  
         
         * This Can Reveal Which Websites Are Being Accessed By Users On The Network, Potentially Used For Targeted Phishing Or Surveillance    


 * **DNS Records :**

     * **PTR Records :**

         * Used for reverse DNS lookups, translating an IP address back to a domain name 

     * **A :**

         * it maps a domain name to an IPv4 address


     * **AAAA :**

         * Similar to A records, but for IPv6 addresses


     * **MX :**

         * Directs email for a domain to a specific mail server

     * **CNAME :**

         * An alias, pointing one domain name to another domain name

# **TCP & UPD :**

 * **TCP (Transmission Control Protocol) :**

     * Provides Reliable, Connection-Oriented Communication With Guaranteed Delivery And Ordered Data Transfer, Making It Suitable For Applications That Require High Reliability And Integrity

     * **Establishing Connections (Three-Way Handshake) :**

         1) Src Host Sends **SYN** Flag To Dst Host 
         2) Dst Host Sends **SYN , ACK** Flags To Src Host
         3) Src Host Sends **ACK Back** Flag To Dst Host  

         * After The Handshake The Real Data Exchange Can Begin

     * **Important Port Numbers That Use TCP :**

         - FTP data (File Transfer Protocol) (20)
         - FTP control (21)
         - SSH (Secure Shell) (22)
         - Telnet (Teletype Network) (23)
         - SMTP (Simple Mail Transfer Protocol) (25)
         - HTTP (HyperText Transfer Protocol) (80)
         - HTTPS (HyperText Transfer Protocol Secure) (443)
         - POP3 (Post Office Protocol version 3) (110)
         - DNS (Domain Name System) (53)  

     * **Common Attack :**

         * **SYN flood :**

             * Attacker Sends A Large Number Of SYN Packets To A Server, Causing It To Become Overwhelmed And Unable To Respond To Legitimate Requests. This Can Lead To A (DoS) attack. 

     * **tcpdump :**

         * Command used for capturing and analyzing network traffic     

         * **tcpdump -i eth0 :** This captures traffic on the network interface named eth0    

         * **tcpdump dst port 80 :** Captures only traffic going to port 80 (usually HTTP traffic)
* * *         


 * **UDP (User Datagram Protocol) :**

     * Offers Low-Overhead, Not Connection-Oriented Communication With Minimal Latency, Making It Suitable For Applications Where Real-Time Performance Is Critical, Even At The Cost Of Occasional Data Loss Or Duplication

     * **Important Port Numbers That Use UDP :**

         - DHCP (Dynamic Host Control Protocol) server (67)
         - DHCP client (68)
         - TFTP (Trivial File Transfer Protocol) (69)
         - SNMP (Simple Network Management Protocol) agent (161)
         - SNMP manger (162)
         - Syslog (System Logging Protocol) (514)
         - DNS (Domain Name System) (53)


     * **Common Attack :**

         * **Smurf :**

             * Attacker Sends A Large Number Of UDP Packets With A Forged Source IP Address To A Broadcast Address, Causing All Machines On The Network To Send A Response To The Forged Address. This Can Also Lead To A DoS Attack.
