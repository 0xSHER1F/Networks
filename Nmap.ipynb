 * **Nmap, short for Network Mapper, is used for network discovery and security auditing**

 * **Basic Syntax :**

     * >> nmap [Options] Target

         * **Target :** 

             * Specifies the target you want to scan. It can be a single IP address, a domain name, a subnet using CIDR notation (192.168.1.0/24), or a file containing a list of targets 

         * **Options :**

             * **-sT :** 

                 * Performs a TCP SYN scan

             * **-sU :** 

                 * Performs a UDP SYN scan  


             * **-p port(s) :**

                 * Specifies the port(s) to scan. You can specify a single port (e.g., 80) or a range of ports separated by a hyphen (-) 


             * **-v :**

                 * Enables verbose output, providing more details about the scan process     


             * **-A :** 

                 * Performs an aggressive scan using various techniques to identify the operating system, services, and versions


             * **-sn :**

                 * Perform Ping Scan , helps identify open ports without revealing your presence as much as a full connection scan


             * **-iL file:**

                 * Reads target IPs from a file


             * **--top-ports (n) :**

                 * instructs Nmap to scan only the top **n** most common ports. By default, Nmap scans the first 1000 ports 


             * **--script= script :** 

                 * Enables scripting capabilities of Nmap. Scripts can gather additional information about detected services on open ports 

                 * Those scripts are located in /usr/share/nmap/scripts


             * **-o fileName:**

                 * Saves the scan results to a specified file    


             * **--oG :** 

                 * Saves the scan results in grepable format, useful for text-based searching and analysis


             * **--oX :**

                 * Saves the scan results in XML format, a machine-readable format for further processing or integration with other tools

             * **-oA :**

                 * Saves the scan results in all available formats (including grepable and XML)
