 * **Hypertext Transfer Protocol (HTTP) It's a set of rules that defines how data is exchanged between devices on the web**

 * **Stateless Protocol : Each request-response interaction is independent, and the server doesn't maintain information about previous interactions with the client**

*** 

 * **HTTP Methods :** 

     * Defines the action the client wants the server to perform on a specific resource

     * **Common Methods :**

         1) **GET :** 

             * Retrieves a resource from the server

         2) **POST :** 

             * Submits data to the server

         3) **DELETE :**

             * Deletes a resource from the server

         4) **HEAD :**

             * Retrieves just the header information of a resource, without the actual content

         5) **OPTIONS :**

             * Discovers the communication options available for a specific resource, like a web page or an API endpoint 

         6) **TRACE :**

             * Used for diagnostic purposes. It allows the client to see what the server receives in the request exactly as it was sent    

             * The server responds with a complete copy of the request message it received, including headers and body (if any)   

             * Often disabled on production servers                


***

 * **HTTP Headers :**

     * **They carry essential information exchanged between client and server during every request and response cycle**


     1) **HTTP Request :**

         * The client initiates the communication by sending an HTTP request message to the server. This message specifies what the client wants from the server  

         * **It Composed Of :**

             1) **Request Line :**

                 * The 1st line of the HTTP request, It specifies the request method (GET- POST ...), the requested resource URL, and and the HTTP version used (HTTP/1.1)

             2) **Request Headers :**

                 * Used to pass additional information about the request to the server , Like :

                     * **User-Agent :** Identifies the software making the request 

                     * **Host :** specifies the hostname of the request

                     * **Accept :** Indicates what type of content the client can accept     

                     * **Referer :** Specifies the URL of the web page that linked to the current request

                     * **Cache-Control :** Defines how and for how long the client can cache the response

                     * **Authorization :** Includes authentication credentials if required to access the resource

                     * **Cookie :** used to send cookies from the client to the server

             3) **Request Body :**

                 * Some methods like POST or PUT may include a message body containing data to be submitted to the server        


     2) **HTTP Response :**  

         * After receiving and processing the request, the server sends an HTTP response message back to the client   

         * **It Composed Of :**

             1) **Status Code Line:**

                 * This line includes the HTTP version, a three-digit status code

             2) **Response Headers :**

                 * Bits of information that browsers and servers use to communicate with each other , Like :

                     1) **Content-Type :** Indicates the type of content being returned        

                     2) **Server :** Identifies the server software that processed the request      

                     3) **Content-Length :** Specifies the size of the response body

                     4) **Cache-Control :** Defines how and for how long the response can be cached by the client

                     5) **Set-Cookie :** is sent by the server to the client to set a cookie

                     6) **Location :** indicates the URL to which to redirect the page

                     7) **Access-Control-Allow-Origin :** indicates which origins can access the pageâ€™s content  


             3) **Response Body :**

                 * Contains the actual content of the web page, such as the HTML and JS code. Once your browser receives all the information needed to construct the web page        

***

 * **Status Code :**

     * **three-digit codes sent by the server inform the client about the outcome of the request** , They are categorized as follow :

         1) **1xx :**

             * **(Informational)** => Request received and processing is continuing

             * **100 Continue :** The server has received the request header and is continuing the process


         2) **2xx :**

             * **(Successful)** => The request was successful  

             * **200 OK :** The request was successful, and the requested data is in the response body
             
             * **201 Created :** The request resulted in the creation of a new resource 

         3) **3xx :** 

             * **(Redirection)** =>  Further action needed to complete the request

             * **301 Moved Permanently :** The requested resource has been permanently moved to a new location

             * **302 Found :** The requested resource can be found at a different URL (temporary or permanent)
             
             
             * **307 Temporary Redirect :** The requested resource has been temporarily moved to a new location

         4) **4xx :**

             * **(Client Error)** => The request contains an error    

             * **404 Not Found :** The requested resource could not be found
             
             * **401 Unauthorized :** The request requires user authentication
             
             * **403 Forbidden :** Access denied


         5) **5xx :**

             * **(Server Error)** => The server failed to fulfill the request

             * **500 Internal Server Error :** The server encountered an internal error and could not complete the request
             
             * **503 Service Unavailable :**  The server is currently unavailable to handle the request (overloaded)

 * **What is The Difference Between Session & Cookie ?**

     * Both used to store user information on websites, but they differ in where that information is stored and how it's accessed 

     * **Cookies :**

         * Stored on the user's device **(client-side)** by the web browser

         * Hold a limited amount of data (usually up to 4KB) and can be accessed by the website that created them and any other website that shares the same domain  

         * **Session Cookies :** Temporary Cookies expire when the user closes the browser window or tab 

         * **Persistent Cookies :** These have an expiration date set in the future, so they remain on the user's device even after the browser is closed. They can be used to remember user preferences across sessions


     * **Session :**

         * Stored on the **server-side**. The server generates a unique session ID and assigns it to the user

         * This ID might be stored in a cookie on the client-side, but the actual data is kept securely on the server

         * Session data typically expires after a period of inactivity on the website 

         * The server can destroy the session or assign a new session ID when the user logs out or becomes inactive

 * **What Is The WebDAV ?**

     * Stands for **Web-Distributed Authoring and Versioning** is a set of extensions built on top of the HTTP protocol

     * It essentially adds functionalities to HTTP that enable collaborative editing and management of files on a web server

         * **Collaborative Editing :** 
         
             * Multiple users can work on the same file simultaneously, similar to how Google Docs functions
             
             * *WebDAV** provides mechanisms for locking files to prevent conflicts and tracks revisions for version control

         * **File Management :**

             * WebDAV allows users to upload, download, delete, move, and rename files on a remote server just like you would with a local file manager
